```python
# Cell 1: Install required libraries
# In Google Colab, some libraries may already be installed,
# but we install them to ensure the environment is ready.

!pip install pandas faker

# Cell 2: Import necessary libraries
# pandas -> used for creating and handling data tables
# faker -> used to generate realistic fake data for HR datasets
# random -> used for generating random values
# datetime -> used for generating date related information

import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

# Initialize Faker object
fake = Faker()

# Cell 3: Create a sample Employee dataset
# This dataset contains basic employee information
# such as ID, Name, Email, Department, Joining Date, and Salary.

num_employees = 20

employees = []

for i in range(num_employees):
    employee = {
        "Employee_ID": 1000 + i,
        "Name": fake.name(),
        "Email": fake.email(),
        "Department": random.choice(["HR", "Finance", "IT", "Marketing", "Operations"]),
        "Job_Title": random.choice(["Manager", "Analyst", "Executive", "Specialist"]),
        "Joining_Date": fake.date_between(start_date='-5y', end_date='today'),
        "Salary": random.randint(30000, 120000)
    }

    employees.append(employee)

# Convert list to DataFrame
employee_df = pd.DataFrame(employees)

# Display sample data
employee_df.head()

# Cell 4: Save Employee dataset as CSV file
# This file can be used for HR analytics or training purposes.

employee_df.to_csv("employees.csv", index=False)

print("employees.csv file created successfully!")

# Cell 5: Create Department dataset
# This dataset contains department information.

departments = {
    "Department_ID": [1,2,3,4,5],
    "Department_Name": ["HR", "Finance", "IT", "Marketing", "Operations"],
    "Manager": [fake.name() for _ in range(5)]
}

department_df = pd.DataFrame(departments)

department_df

# Cell 6: Save Department dataset to CSV

department_df.to_csv("departments.csv", index=False)

print("departments.csv file created successfully!")

# Cell 7: Create Attendance dataset
# This dataset tracks employee attendance status.

attendance_records = []

for i in range(50):
    record = {
        "Employee_ID": random.choice(employee_df["Employee_ID"]),
        "Date": fake.date_between(start_date='-30d', end_date='today'),
        "Status": random.choice(["Present", "Absent", "Work From Home", "Leave"])
    }

    attendance_records.append(record)

attendance_df = pd.DataFrame(attendance_records)

attendance_df.head()

# Cell 8: Save Attendance dataset to CSV

attendance_df.to_csv("attendance.csv", index=False)

print("attendance.csv file created successfully!")

# Cell 9: Create Recruitment dataset
# This dataset represents job applicants and their application status.

applicants = []

for i in range(15):
    applicant = {
        "Applicant_ID": 5000 + i,
        "Name": fake.name(),
        "Email": fake.email(),
        "Position_Applied": random.choice(["Data Analyst", "HR Executive", "Software Engineer", "Marketing Manager"]),
        "Application_Date": fake.date_between(start_date='-60d', end_date='today'),
        "Status": random.choice(["Shortlisted", "Interviewed", "Rejected", "Hired"])
    }

    applicants.append(applicant)

recruitment_df = pd.DataFrame(applicants)

recruitment_df.head()

# Cell 10: Save Recruitment dataset to CSV

recruitment_df.to_csv("recruitment.csv", index=False)

print("recruitment.csv file created successfully!")

# Cell 11: Create Payroll dataset
# This dataset contains salary and bonus details of employees.

payroll_records = []

for emp in employee_df["Employee_ID"]:
    record = {
        "Employee_ID": emp,
        "Base_Salary": random.randint(30000, 120000),
        "Bonus": random.randint(2000, 15000),
        "Deductions": random.randint(1000, 5000),
        "Net_Salary": random.randint(28000, 115000)
    }

    payroll_records.append(record)

payroll_df = pd.DataFrame(payroll_records)

payroll_df.head()

# Cell 12: Save Payroll dataset to CSV

payroll_df.to_csv("payroll.csv", index=False)

print("payroll.csv file created successfully!")
```
