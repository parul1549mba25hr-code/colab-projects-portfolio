```python
import pandas as pd
df = pd.read_csv("sales-revenue-business-dataset.csv")
print(df)
```

```python
# DATA PRE-PROCESSING
df = df.drop('column1' , axis=1)
```

```python
df.columns
```

```python
df.isnull()
```

```python
# missing the filling values
# filling means replace the NaN values with 0
df.fillna(0)
```

```python
# dropna (will remove the entire row with NaN value)
df = df.dropna()
```

```python
df
```

```python
df.to_csv("cleaned_sales_dataset.csv")
```

```python
# feature engineering
df = df.drop("unnamed: 0", axis = 1)
```

```python
df.columns
```

```python
# create a new column
df['profit] = df['revenue']-df['cost']
```

```python
df.head()
# .head() shows top 5 rows
# .tail() shows bottom 5 rows
```

```python
# create profit margin column
df['profit_margin'] = df['profit']/df['revenue']
```
