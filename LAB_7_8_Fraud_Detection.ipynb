```python
# Step 1: Import required libraries
# pandas is used for data handling
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Step 2: Create the transaction dataset

# This dataset represents basic customer transaction behavior
# transaction_id -> unique transaction number
# customer       -> customer name
# amount         -> transaction amount
# location       -> place from where transaction was made
# transactions_in_1hr -> number of transactions made by same customer in 1 hour
# fraud          -> target column
#                  0 = normal transaction
#                  1 = suspicious / fraudulent transaction

data = {
    "transaction_id": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "customer": ["Amit", "Riya", "Sohan", "Neha", "Kiran", "Amit", "Riya", "Sohan", "Neha", "Kiran"],
    "amount": [2000, 5000, 300, 70000, 4500, 150000, 6000, 250, 90000, 4000],
    "location": ["Delhi", "Mumbai", "Delhi", "Jaipur", "Mumbai", "Unknown", "Mumbai", "Delhi", "Unknown", "Mumbai"],
    "transactions_in_1hr": [1, 2, 1, 3, 2, 6, 2, 1, 5, 1],
    "fraud": [0, 0, 0, 1, 0, 1, 0, 0, 1, 0]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display dataset
df
# Step 3: Inspect the data

# head() shows first 5 rows
print("First 5 rows:")
print(df.head())

# shape shows number of rows and columns
print("\nShape of dataset:")
print(df.shape)

# info() shows column details and data types
print("\nDataset information:")
print(df.info())
# Step 4: Understand the business problem

# Fraud detection means identifying unusual or suspicious transactions
# Example signals:
# very high amount
# too many transactions in a short time
# unknown or suspicious location

print("Fraud Detection Business Case:")
print("We want to identify transactions that look abnormal or suspicious.")

# Step 5: Find transactions with very high amount

# This is a simple business rule:
# if amount is greater than 50000, mark it as suspicious for review

high_amount_transactions = df[df["amount"] > 50000]

print("Transactions with High Amount:")
print(high_amount_transactions)
# Step 6: Find transactions with too many transactions in 1 hour

# Another business rule:
# if the customer makes more than 4 transactions in 1 hour,
# it may indicate suspicious behavior

frequent_transactions = df[df["transactions_in_1hr"] > 4]

print("Transactions with High Frequency in 1 Hour:")
print(frequent_transactions)

# Step 7: Find transactions from suspicious or unknown location

# If the transaction comes from an unusual location,
# it may need further verification

unknown_location_transactions = df[df["location"] == "Unknown"]

print("Transactions from Unknown Location:")
print(unknown_location_transactions)

# Step 8: Create a simple rule-based fraud flag

# We will mark a transaction as suspicious if:
# amount > 50000
# OR transactions_in_1hr > 4
# OR location is Unknown

df["predicted_fraud"] = (
    (df["amount"] > 50000) |
    (df["transactions_in_1hr"] > 4) |
    (df["location"] == "Unknown")
).astype(int)

# Step 9: Compare actual fraud vs predicted fraud

# fraud           -> actual label
# predicted_fraud -> fraud identified by our simple rule-based logic

print("Actual vs Predicted Fraud:")
print(df[["transaction_id", "customer", "fraud", "predicted_fraud"]])

# Step 10: Check model/rule accuracy manually

# This checks how many fraud predictions matched the actual fraud values

correct_predictions = (df["fraud"] == df["predicted_fraud"]).sum()
total_transactions = len(df)
accuracy = correct_predictions / total_transactions

print("Correct Predictions:", correct_predictions)
print("Total Transactions:", total_transactions)
print("Accuracy:", accuracy)


# Step 11: Show only the suspicious transactions identified by the system

suspicious_transactions = df[df["predicted_fraud"] == 1]

print("Suspicious Transactions Identified by the System:")
print(suspicious_transactions)

# Step 12: Business interpretation

print("Business Interpretation:")
print("The system identified suspicious transactions based on high amount,")
print("high transaction frequency, or unknown location.")
print("In real companies, such transactions are sent for manual review")
print("or blocked temporarily for customer safety.")
```
